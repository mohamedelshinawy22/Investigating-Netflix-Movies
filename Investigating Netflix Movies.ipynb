```python
import pandas as pd

# Load the dataset
df = pd.read_csv('netflix_data.csv')

# Convert 'release_year' to numeric (or correct column name if different)
df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce')

# Filter for movies from the 1990s
df_1990s = df[(df['release_year'] >= 1990) & (df['release_year'] < 2000)]

# Convert 'duration' to numeric, forcing errors to NaN
df_1990s['duration'] = pd.to_numeric(df_1990s['duration'], errors='coerce')

# Find the most frequent movie duration in the 1990s
most_frequent_duration = df_1990s['duration'].mode()

# Convert the most frequent duration to an integer
if not most_frequent_duration.empty:
    duration = int(most_frequent_duration[0])
else:
    duration = None

# Count the number of short action movies released in the 1990s
short_action_movies = df_1990s[(df_1990s['genre'] == 'Action') & (df_1990s['duration'] < 90)]
short_movie_count = len(short_action_movies)

# Print the results
if duration is not None:
    print(f"Most frequent movie duration in the 1990s: {duration} minutes")
else:
    print("No movie duration data available for the 1990s")

print(f"Number of short action movies released in the 1990s: {short_movie_count}")
```
